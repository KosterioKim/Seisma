```python
import matplotlib.pyplot as plt
```

```python
def read_data(file_path):
   data = []
   with open(file_path, 'r') as file:
       for line in file:
           if line.startswith('#') or not line.strip():
               # Пропускаем строки с комментариями и пустые строки
               continue
           # Разделяем строку по пробелам и преобразуем в числа
           x, y, z, col, row = map(float, line.split())
           data.append((x, y, z, int(col), int(row)))
   return data
```

```python
# Пример использования функции
file_path = 'C:/HV/Seismic/datas/Surface_attribute_AmplitudeContrast_HalfEnergy_TWT.txt'
scattered_data = read_data(file_path)
```

```python
x_coords = [item[0] for item in scattered_data]
y_coords = [item[1] for item in scattered_data]
values = [item[2] for item in scattered_data]
zero_coordinats = []
for i in range(len(values)):
   if values[i] <=0:
       zero_coordinats.append([x_coords[i], y_coords[i]])
```

```python
# Создание карты с использованием scatter plot
plt.scatter(x_coords, y_coords, c=values, cmap='viridis')

# Выделение определенных ячеек
for cell in zero_coordinats:
   # Выделение красным цветом и увеличение размера
   plt.scatter(*cell, color='red', edgecolor='black', s=100)

plt.colorbar()  # Показывает шкалу значений
plt.xticks([x_coords[0], int((x_coords[-1] - x_coords[0])//2 + x_coords[0]), x_coords[-1]])
plt.yticks([y_coords[0], int((y_coords[-1] - y_coords[0])//2 + y_coords[0]), y_coords[-1]])
# Настройка заголовков и меток осей
plt.title('Карта значений')
plt.xlabel('Inlines')
plt.ylabel('Xlines')

# Отображение карты
plt.show()

if len(zero_coordinats)==0:
   print('тест пройден успешно')
else:
   print('Убедитесь, что поверхностный сейсмический атрибут рассчитан корректно')
```
