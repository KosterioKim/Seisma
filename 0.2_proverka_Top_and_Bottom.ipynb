```python
import matplotlib.pyplot as plt
```

```python
def read_data(file_path):
   data = []
   with open(file_path, 'r') as file:
       for line in file:
           if line.startswith('#') or not line.strip(): # Пропускаем строки с комментариями и пустые строки
               continue
           # Разделяем строку по пробелам и преобразуем в числа
           x, y, z, col, row = map(float, line.split())
           data.append([x, y, z, int(col), int(row)])
   return data
```

```python
def show_pic(datas, max, min):
   matrix_map = []
   matrix_map.append([])
   j = 0
   chek = datas[0][1]
   zero = []
   for i in range(len(datas)):
       if datas[i][2] >= 0:
           zero.append([datas[i][0], datas[i][1]])
       if chek == datas[i][1]:
           matrix_map[j].append(datas[i][2] * -1)
       else:
           chek = datas[i][1]
           matrix_map.append([])
           j += 1
           matrix_map[j].append(datas[i][2] * -1)


   plt.figure(figsize=(5, 5), dpi=150)
   plt.imshow(matrix_map, vmin=min, vmax=max, cmap='seismic')
   plt.xlabel('inline')
   plt.ylabel('xline')
   plt.grid(ls=':', alpha=.5)
   plt.colorbar(shrink=0.5)
   plt.show()
   if len(zero) == 0:
       print('тест пройден успешно')
   else:
       print('Значения времени TWT должны быть положительными')
       print(zero)
```

```python
file_path = 'C:/HV/Seismic/datas/TWT_Top_U1.txt'
scattered_data = read_data(file_path)
show_pic(scattered_data, 2300, 2200 )
```

```python
show_pic(read_data('C:/HV/Seismic/datas/TWT_Bottom_U1.txt'), 2400, 2300)
```
